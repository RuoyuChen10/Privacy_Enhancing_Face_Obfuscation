```python
import cv2
import torch
import argparse
import os
import numpy as np
import json
import sys
sys.path.append("../")

from models import BranchedTinyAttr
from utils import load_yaml, read_img
from tqdm import tqdm
import random
```

```python
device = "cuda:0"

cfg = load_yaml("/home/lsf/桌面/MaskFaceGAN/config.yml")
model = BranchedTinyAttr(cfg.MODELS.CLASSIFIER)

model.set_idx_list(attributes=['male', "young"])

model.to(device)
```

```python
gender = ["Female", "Male"]
age = ["Old", "Young"]
```

```python
for i in range(1,10):
    path1 = os.path.join("image-4.4/real", str(i)+".jpg")
    path2 = os.path.join("image-4.4/fake", str(i)+".jpg")
    
    image_1 = read_img(path1)
    image_2 = read_img(path2)
    
    pred1 = torch.sigmoid(model(image_1.to(device)))
    pred2 = torch.sigmoid(model(image_2.to(device)))
    
    pred1 = (pred1>0.5).int()
    pred2 = (pred2>0.5).int()
    
    des1 = "Attribute: " + gender[pred1[0][0]] + ", " + age[pred1[0][1]]
    des2 = "Attribute: " + gender[pred2[0][0]] + ", " + age[pred2[0][1]]
    
    print("Image {}, Left {}, Right {}.".format(i, des1, des2))
```
